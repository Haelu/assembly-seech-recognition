```python
import sys
import time
import requests
 
def read_file(filename, chunk_size=5242880):
    with open(filename, 'rb') as _file:
        while True:
            data = _file.read(chunk_size)
            if not data:
                break
            yield data
 
headers = {'authorization': "cfcf3caf09a2463899e902628d9253db"}
response = requests.post('https://api.assemblyai.com/v2/upload',
                         headers=headers,
                         data=read_file('sample.wav'))

print(response.json())
```

```python
endpoint = "https://api.assemblyai.com/v2/transcript"

json = {
  "audio_url": "https://cdn.assemblyai.com/upload/26af101e-f99d-4fce-94ff-bfe768fe9002"
}

headers = {
    "authorization": "cfcf3caf09a2463899e902628d9253db",
    "content-type": "application/json"
}

response = requests.post(endpoint, json=json, headers=headers)

print(response.json())
```

```python

transcript_id=response.json()['id']
endpoint = "https://api.assemblyai.com/v2/transcript/"+transcript_id

headers = {
    "authorization": "cfcf3caf09a2463899e902628d9253db",
}

response = requests.get(endpoint, headers=headers)

print(response.json())
```
